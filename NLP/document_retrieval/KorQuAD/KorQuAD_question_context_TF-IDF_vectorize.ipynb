```python
import os, sys
import pandas as pd
import numpy as np

from konlpy.tag import Twitter
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
data = pd.read_csv('../../data/KorQuAD/KorQuAD_v1.0_train_preprocess.txt', sep='\t', header=(0))
```

```python
data
```

```python
question_corpus = data[(pd.notnull(data['question'].values) & pd.notnull(data['context'].values))]['question'].values
context_corpus = data[(pd.notnull(data['question'].values)  & pd.notnull(data['context'].values))]['context'].values
```

```python
question_corpus = question_corpus.tolist()
context_corpus = context_corpus.tolist()
```

```python
tokenizer = Twitter()
```

```python
def twitter_tokenizer(sentence):
    word_array = []
    for eachPart in tokenizer.pos(sentence):
        word = eachPart[0]
        part = eachPart[1]
        word_array.append(word + '/' + part)
        
    return word_array
```

```python
vectorizer = TfidfVectorizer(tokenizer=twitter_tokenizer)
question_tfidf_vectors = vectorizer.fit_transform(question_corpus)
context_tfidf_vectors = vectorizer.fit_transform(context_corpus)
```
